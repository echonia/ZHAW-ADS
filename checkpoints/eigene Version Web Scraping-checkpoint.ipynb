```python
# printen der image Tags
imTags=soup.find_all('img',{'class':'product-image-photo'})
print(imTags[0])
```

```python
# single one - src Tag entscheidend...
src = imTags[0:0]['src']
print(src)
```

```python
# adden einen url (mit itr+1,  weil wir mehrere Seiten wollen). Definieren nicht extra function.
# src ist HTML-Tag f√ºr Bilder-Links
# f vor String ist n√∂tig, weil sonst nur eine Seite (f = beh√§lt Sematik bei mehr Zeilen und und...  ??)

for itr in range(100):
    try:
        url = f"https://www.koala.ch/en/shoes/sneakers.html?p={itr+1}"
        r = requests.get(url)
        h = r.content
        soup = BeautifulSoup(h, 'lxml')
        imTags=soup.find_all('img',{'class':'product-image-photo'})
        
        for i in imTags:
            sneakers.append(i['src'])
    except:
        pass
    # break funktioniert aus irgend einem grund nicht, deshalb pass
```

## Liste speichern und Bilder in Ordner 

```python
sneakers = list(dict.fromkeys(sneakers))
try:
    sneakers2=sneakers[:1000]
except:
    sneakers2=sneakers
for i,v in enumerate(sneakers2):
    with open('sneakers/'+str(i)+'.jpg','wb') as f:
        im=requests.get(v)
        f.write(im.content)
```

## Vorgang f√ºr Ballerinas

```python
# Ordner kreieren, falls nicht vorhanden f√ºr Sneakers
dir = os.path.join(r'C:\Users\benip\OneDrive\ZHAW\2 Term\Di, 12.50-16.45 (ADS) Applied Data Science\Projekt ADS FS 2022','ballerinas')
if not os.path.exists(dir):
    os.mkdir(dir)
    

# liste initialisieren
ballerinas = []
r=0
```

```python
# adden einen url (mit itr+1,  weil wir mehrere Seiten wollen). Definieren nicht extra function.
# src ist HTML-Tag f√ºr Bilder-Links
# f vor String ist n√∂tig, weil sonst nur eine Seite (f = xxxxxx ??)

for itr in range(100):
    try:
        url = f"https://www.koala.ch/en/shoes/ballerinas.html?p={itr+1}"
        r = requests.get(url)
        h = r.content
        soup = BeautifulSoup(h, 'lxml')
        imTags=soup.find_all('img',{'class':'product-image-photo'})
        
        for i in imTags:
            ballerinas.append(i['src'])
    except:
        pass
    # break funktioniert aus irgend einem grund nicht, deshalb pass
```

```python
# allf√§llige Duplikate aus der Liste entfernen
ballerinas = list(set(ballerinas))

    
# √∂ffne Ballerinas + string hinter image mit Nummer, bspw. 1 + 
# wb = write and binary --> must be used when writing non-text files like images
for i,v in enumerate(ballerinas):
    with open('ballerinas/'+str(i)+'.jpg','wb') as f:
        im=requests.get(v)
        f.write(im.content)
```
