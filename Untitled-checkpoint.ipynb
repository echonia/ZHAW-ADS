```python
# adden einen url (mit itr+1,  weil wir mehrere Seiten wollen). Definieren nicht extra function.
# src ist HTML-Tag f√ºr Bilder-Links
# f vor String ist n√∂tig, weil sonst nur eine Seite (f = beh√§lt Sematik bei mehr Zeilen und und...  ??)

for itr in range(100):
    try:
        url = f"https://www.koala.ch/en/shoes/sneakers.html?p={itr+1}"
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'lxml')
        imTags=soup.find_all('img',{'class':'product-image-photo'})
        
        for i in imTags:
            sneakers.append(i['data-po-cmp-src'])
    except:
        pass
```

#### Liste speichern und Bilder in Ordner 

```python
sneakers = list(dict.fromkeys(sneakers)) # removing the duplicate src links
try:
    sneakers2=sneakers[:1000] # why this part?
except:
    sneakers2=sneakers
for i,v in enumerate(sneakers2):
    with open('sneakers/'+str(i)+'.jpg','wb') as file: # opening images as JPG Files
        im=requests.get(v) # warum hier nochmals get request?
        file.write(im.content)
```
