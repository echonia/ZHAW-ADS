```python
# teile der Funktion
r = requests.get("https://www.koala.ch/de/shoes/sneakers.html?p=")
print(r)
```

```python
h = r.content
print(h)
```

```python
soup=BeautifulSoup(h, 'lxml') 
print(soup.prettify())
# Text ist nach HTML Tags gegliedert, etwas unsch√∂n
# mit Prettify Funktion ein M√ºh besser
```

```python
# printen der image Tags
imTags=soup.find_all('img',{'class':'product-image-photo'})
print(imTags[2])
```

```python
# single one - src Tag entscheidend...
src = imTags[2]['data-po-cmp-src']
print(src)
```

```python
# adden einen url (mit itr+1,  weil wir mehrere Seiten wollen). Definieren nicht extra function.
# f vor String ist n√∂tig, damit itr+1 erkannt wird als separater Teil
# data-po-cmp-src ist HTML-Tag f√ºr Bilder-Links
# pass ist n√∂tig, da wir, wenn eins nicht funktioniert, keinen Stop (break) wollen, sondern nur eins weiter

for itr in range(100):
    try:
        url = f"https://www.koala.ch/en/shoes/sneakers.html?p={itr+1}"
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'lxml')
        imTags=soup.find_all('img',{'class':'product-image-photo'})
        
        for i in imTags:
            sneakers.append(i['data-po-cmp-src'])
    except:
        pass
```

#### Liste speichern und Bilder in Ordner 

```python
# allf√§llige Duplikate aus der Liste entfernen
sneakers = list(dict.fromkeys(sneakers)) 
```

```python
# √∂ffne Heels + string hinter image mit Nummer, bspw. 1 + Endung jpg
# wb = write and binary --> must be used when writing non-text files like images
for i,v in enumerate(sneakers):
    with open('sneakers/'+str(i)+'.jpg','wb') as file: 
        im=requests.get(v)
        file.write(im.content)
```
